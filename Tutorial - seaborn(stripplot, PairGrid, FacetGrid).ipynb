# seaborn - stripplot, PairGrid, FacetGrid

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

sns.set(style="darkgrid")
```

## stripplot

```python

tips = sns.load_dataset('tips')
tips.head(5)
```

```python

sns.stripplot(x='day', y='total_bill', data=tips, marker='D', size=10)

plt.figure()
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', palette='autumn', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', palette='autumn', split=True, edgecolor='gray',
              alpha=0.2)
```

```python

sns.stripplot(x='tip', y='day', data=tips, jitter=True)
sns.violinplot(x='tip', y='day', data=tips, palette='coolwarm', color='l')
```

## PairGrid  
  
seaborn.PairGrid is a class. So it need to initialize. The example is show below:  
http://seaborn.pydata.org/generated/seaborn.PairGrid.html?highlight=pairgrid

```python

iris = sns.load_dataset('iris')

x = sns.PairGrid(iris, size=3)
x = x.map(plt.scatter)
```

```python

x = sns.PairGrid(iris, hue='species', palette='husl')
x = x.map_diag(plt.hist, histtype='step', linewidth=3) # only diagonal
x = x.map_offdiag(plt.scatter) # not diagonal
x = x.add_legend() # add legend
```

```python

x = sns.PairGrid(iris, hue='species', palette='husl', vars=['petal_length','petal_width'])
x = x.map_offdiag(plt.scatter)
x = x.map_diag(plt.hist, edgecolor='w')
x = x.add_legend() # add legend
```

```python

x = sns.PairGrid(iris, hue='species', palette='husl', x_vars=['petal_length','petal_width'],
                 y_vars=['sepal_length','sepal_width'])
x = x.map(plt.scatter)
x = x.add_legend() # add legend
```

```python

x = sns.PairGrid(iris, palette='husl')
x = x.map_upper(plt.scatter)
x = x.map_diag(plt.hist, edgecolor='w')
x = x.map_lower(sns.kdeplot)
x = x.add_legend() # add legend
```

## FacetGrid

```python

x = sns.FacetGrid(tips, row='smoker', col='time')
x = x.map(plt.hist, 'total_bill',edgecolor='w')
```

```python

x = sns.FacetGrid(tips, row='smoker', col='time', hue='sex')
# x = x.map(plt.scatter, 'total_bill','tip')
x = x.map(sns.regplot, 'total_bill', 'tip').add_legend()
# x.add_legend()
```

```python

x = sns.FacetGrid(tips, col='day')
x = x.map(sns.boxplot, 'time', 'total_bill')
# x.add_legend()
```
