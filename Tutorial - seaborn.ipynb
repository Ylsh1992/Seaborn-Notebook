# Tutorial - seaborn

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

sns.set(style="darkgrid")
```

## KDE plot

```python

mean = [0,0]
cov = [[0.2,0],[0,3]]

x,y=np.random.multivariate_normal(mean,cov,size=40).T
x.shape
sns.kdeplot(x, shade=True)

plt.figure()
sns.kdeplot(x,y, n_levels=18, cmap='coolwarm')
```

```python

iris = sns.load_dataset('iris')

setosa = iris.loc[iris.species == 'setosa']
versicolor = iris.loc[iris.species == 'versicolor']
```

```python

sns.kdeplot(setosa.petal_length, setosa.petal_width, cmap='coolwarm')

sns.kdeplot(versicolor.petal_length, versicolor.petal_width, cmap='RdBu')
```

## Joint plot 

```python

tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python

sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python

sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='reg')
plt.figure()
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='hex')
# kind = reg, hex kde....
```
