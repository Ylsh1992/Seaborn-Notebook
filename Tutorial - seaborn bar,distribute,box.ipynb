# Tutorial - seaborn bar,distribute,box

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

sns.set(style="darkgrid")
```

## barplot

```python

tips = sns.load_dataset('tips')
tips.head(5)
# tips.describe()
```

```python

sns.barplot(x='day', y='tip', data=tips)

plt.figure()
sns.barplot(x='day', y='total_bill', data=tips, palette='coolwarm')
```

```python
from numpy import mean
plt.figure(figsize=(7,7))
# sns.set_style("whitegrid")
sns.barplot(x='day', y='tip', data=tips, hue='sex',  palette='winter_r',ci=55,capsize=0.2)

plt.figure(figsize=(7,7))
sns.barplot(x='day', y='tip', data=tips, hue='sex',  color='blue' , saturation=0.1 ,ci=55,capsize=0.2)
```

## dist plot (distribute)

```python
# numpy
num = np.random.randn(150)
sns.distplot(num,color='red')
# print(num)
```

```python
# pandas series
label_dist = pd.Series(num, name="Variable x")
sns.distplot(label_dist, vertical=True)

plt.figure()
sns.distplot(label_dist, hist=False)
```

## Box Plot

```python
tips.head(5)

sns.boxplot(x=tips['size'])

plt.figure()
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex',palette='husl')
```

```python
# using iris dataset

iris = sns.load_dataset('iris')
iris.head(5)
```

```python

# sns.set(style="darkgrid")
sns.boxplot(data=iris, palette='coolwarm')   # orient= h or v  
sns.swarmplot(data=iris, color='0.5')

plt.figure()
sns.boxplot(x='species', y='petal_length',data=iris,palette='coolwarm')
```
